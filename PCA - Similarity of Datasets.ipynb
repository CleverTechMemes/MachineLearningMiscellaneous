```python
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

# Defining the StandardScaler
scaler = StandardScaler()
# Calculating two major compontents
pca = PCA(n_components=2)

# First scaling the data
X_2019_Scaled = scaler.fit(data_2019).transform(data_2019)
X_2020_Scaled = scaler.fit(data_2020).transform(data_2020)
```

```python
# Let's calculate the covariance matrix
x_2019_covarianceMatrix = np.cov(np.transpose(np.array(X_2019_Scaled)))
x_2020_covarianceMatrix = np.cov(np.transpose(np.array(X_2020_Scaled)))

# Calculating the principal components in our data
pca_2019 = pca.fit(x_2019_covarianceMatrix).transform(x_2019_covarianceMatrix)
pca_2020 = pca.fit(x_2020_covarianceMatrix).transform(x_2020_covarianceMatrix)
```

```python

# Let's vertically add the two prinicapl component to make a single vector 
df_2019_pca = pd.DataFrame(pca_2019, columns= ['comp1','comp2'])
df_2019_pca_vertically_apppended = df_2019_pca['comp1'].append(df_2019_pca['comp2']).reset_index(drop=True)

df_2020_pca = pd.DataFrame(pca_2020, columns= ['comp1','comp2'])
df_2020_pca_vertically_apppended = df_2020_pca['comp1'].append(df_2020_pca['comp2']).reset_index(drop=True)

# Similarity between 2019 and 2020 data

print('correlation of data of 2019 and 2020 is {}'
      .format(np.corrcoef(df_2019_pca_vertically_apppended,df_2020_pca_vertically_apppended)[0,1]))
```
